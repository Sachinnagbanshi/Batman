```python
import mysql.connector as m
mydb=m.connect(host='localhost',user='root',password='sachin')
x=mydb.cursor()
```

```python
def insert(x):
    import pandas as pd
    x.execute(f"use {input('Enter Database Name: ')}")
    table_name=input("Enter Table Name: ")
    x.execute(f"Select * from {table_name};")
    df=pd.DataFrame(x.fetchall(),columns=x.column_names)
    columns=df.columns
    no_of_rows=int(input("How many rows do you want to enter?: "))
    for i in range(0,no_of_rows):
        detail=()
        for j in range(len(columns)):
            if df[columns[j]].dtype=="int64":
                detail+=(int(input(f"Enter the 'Int' value of {columns[j]}: ")),)
            elif df[columns[j]].dtype=="O":
                detail+=(input(f"Enter the 'char' value of {columns[j]}: "),)
            elif df[columns[j]].dtype=="float64":
                detail+=(float(input(f"Enter the 'float' value of {columns[j]}: ")),)
        x.execute(f"Insert into {table_name} values {detail};")
    x.execute('set autocommit=1')
                
```

```python
insert(x)
```
