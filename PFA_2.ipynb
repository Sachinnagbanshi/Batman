```python
#loading assignment

df2=pd.read_excel('python assignment.xlsx',sheet_name=1)
df2
```

```python
df2.info()
```

```python
#Question 1 ---> Finding the no. of duplicate values in AGMTNO column

print(df2.AGMTNO.duplicated().sum())

if df2.AGMTNO.duplicated().sum()!=0:
    print('There is duplicate values.')
else:
    print('There is no duplicate values.')
```

```python
#Hence, no duplicate values found.
```

```python
#Question 2----> Changing the datdf2['Agreement Date'] = pd.to_datetime(df2['Agreement Date'], format='%Y-%m-%d')e format to yyyy/mm/dd

df2['Agreement Date'] = pd.to_datetime(df2['Agreement Date'], format='%Y-%m-%d')
df2
```

```python
#Question 3------> Concatenate customer(add1,add2,add3) to form a single column application address

df2['Applicant Address'] = df2[df2.columns[4:7]].apply(lambda x: ','.join(x.dropna().astype(str)),axis=1)
df2
```

```python
#Question 4:-  (i) Clear 0 and '-' from CONTACT_NO:


df2['CONTACT_NO'].replace('-',0,inplace=True)  #replacing '-' to 0
df2.fillna(0,inplace=True) #filling null values with 0
df2 = df2[df2['CONTACT_NO'] != 0] #removing rows which have 0 in CONTACT_NO
df2
```

```python
#Question 4 (ii)-----> checking the lenght of the contact no., if its less then 10 then show its values corresponding to AGMNTO

for i in df2['CONTACT_NO']:
    if len(i)!=10:
        print(df2['AGMTNO'][df2['CONTACT_NO']==i])
```

```python
#Question 5-----> find '*' in Chassis no. and its corresponding AGMTNO values.

for i in df2['CHASSIS']:
    if '*' in i:
        print(df2['AGMTNO'][df2['CHASSIS']==i])
```

```python
#Question 6------> Round the Total column

df2['Total']=df2['Total'].apply(np.round)
df2
```
